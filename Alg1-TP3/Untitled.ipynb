```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
text_file = open("times.txt", "r")
lines = text_file.read().split('\n')
```

```python
size = []
time = [] 
for string in lines:
    a = string.split(',')
    #print(a[1])
    size.append(a[0])
    time.append(a[-1])
time.remove('')
size.remove('')
```

```python
size = [float(i) for i in size]
time = [float(i) for i in time]
#size = size[0:200] 
#time = time[0:200]
#size = [i*i for i in size]
```

```python
time = time[202: ]
size = size[202: ]
```

```python
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))

#plt.scatter(size,time)
plt.plot( size, time, linestyle='none', marker='o')

#ply.ylabel
```

```python
# library & dataset
import seaborn as sns
sns.set(style="darkgrid") 
 #time = [i/1000 for i in time]
# use the function regplot to make a scatterplot
ax = sns.regplot(x=size, y=time, fit_reg=False)
ax.set(xlabel='Número de arestas + Número de vértices', ylabel='Tempo de execução (s)')
#sns.plt.show()
```
